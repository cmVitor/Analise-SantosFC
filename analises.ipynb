 ## Importar tabela para an√°lise

```python
import pandas as pd

tabela = pd.read_excel("BRA.xlsx")

display(tabela)
```

## Analisando vit√≥ria de mandantes e visitantes

Gr√°fico de barras

```python
import matplotlib.pyplot as plt

contagemResultados = tabela['Res'].value_counts()

contagemResultados.plot(kind='bar', color=['blue', 'green', 'red'])

# Adicionando t√≠tulo e r√≥tulos
plt.title('Distribui√ß√£o dos Resultados')
plt.xlabel('Resultado')
plt.ylabel('Frequ√™ncia')

#Mostrar a contagem
print(contagemResultados)
# Mostrando o gr√°fico
plt.show()
```

Gr√°fico de Pizza

```python
contagemResultados.plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'green', 'red'], labels=['Home', 'Draw', 'Away'])

# Adicionando t√≠tulo
plt.title('Distribui√ß√£o dos Resultados')

# Mostrando o gr√°fico
plt.ylabel('')  # Remove o r√≥tulo do eixo Y
plt.show()
```

## Filtrando os jogos do Santos

```python
jogos_santos = tabela[(tabela['Home'] == 'Santos') | (tabela['Away'] == 'Santos') ] 

display(jogos_santos)
```

## Analisando os resultados do santos

```python
resultados_santos = jogos_santos.apply(
    lambda row: 'Vit√≥ria' if (row['Home'] == 'Santos' and row['Res'] == 'H') or (row['Away'] == 'Santos' and row['Res'] == 'A') 
    else ('Derrota' if (row['Home'] == 'Santos' and row['Res'] == 'A') or (row['Away'] == 'Santos' and row['Res'] == 'H')
    else 'Empate'), axis=1).value_counts()

# Mostrar a contagem
print(resultados_santos)

# Plotar um gr√°fico de barras
resultados_santos.plot(kind='bar', color=['green', 'red', 'gray'])

plt.title('Resultados do Santos')
plt.xlabel('Resultado')
plt.ylabel('Frequ√™ncia')
plt.show()
```

Analisando retrospecto do Santos como mandante e visitante

```python
# Performance em casa vs fora
performance_casa_fora = jogos_santos.groupby(['Home', 'Res']).size().unstack(fill_value=0)
performance_casa_fora.loc['Santos'].plot(kind='bar', color=['red', 'gray', 'green'])

plt.title('Performance do Santos em Casa')
plt.xlabel('Resultado')
plt.ylabel('Frequ√™ncia')
plt.show()

# Contar os resultados do Santos em casa
resultados_casa = jogos_santos[jogos_santos['Home'] == 'Santos'].apply(
    lambda row: 'Vit√≥ria' if row['Res'] == 'H' else ('Derrota' if row['Res'] == 'A' else 'Empate'), axis=1).value_counts()

# Exibir resultados
print("Resultados do Santos em Casa:")
print(f"Vit√≥rias: {resultados_casa.get('Vit√≥ria', 0)}")
print(f"Derrotas: {resultados_casa.get('Derrota', 0)}")
print(f"Empates: {resultados_casa.get('Empate', 0)}\n")

# Performance fora de casa
performance_casa_fora = jogos_santos.groupby(['Away', 'Res']).size().unstack(fill_value=0)
performance_casa_fora.loc['Santos'].plot(kind='bar', color=['green', 'gray', 'red'])

plt.title('Performance do Santos Fora de Casa')
plt.xlabel('Resultado')
plt.ylabel('Frequ√™ncia')
plt.show()

# Contar os resultados do Santos fora de casa
resultados_fora = jogos_santos[jogos_santos['Away'] == 'Santos'].apply(
    lambda row: 'Vit√≥ria' if row['Res'] == 'A' else ('Derrota' if row['Res'] == 'H' else 'Empate'), axis=1).value_counts()

# Exibir resultados
print("Resultados do Santos Fora de Casa:")
print(f"Vit√≥rias: {resultados_fora.get('Vit√≥ria', 0)}")
print(f"Derrotas: {resultados_fora.get('Derrota', 0)}")
print(f"Empates: {resultados_fora.get('Empate', 0)}")
```

## Analisando evolu√ß√£o de resultados ao longo do tempo

```python
# Converter a coluna de data para datetime, se ainda n√£o tiver sido feito
jogos_santos.loc[:, 'Date'] = pd.to_datetime(jogos_santos['Date'])

# Filtrar as vit√≥rias do Santos
vitorias_santos = jogos_santos[((jogos_santos['Home'] == 'Santos') & (jogos_santos['Res'] == 'H')) |
                               ((jogos_santos['Away'] == 'Santos') & (jogos_santos['Res'] == 'A'))]

# Agrupar por ano e contar as vit√≥rias
vitorias_por_ano = vitorias_santos.groupby(vitorias_santos['Date'].dt.year).size()

# Plotando o gr√°fico de linha
vitorias_por_ano.plot(kind='line', marker='o', color='green', figsize=(10, 6))

# Adicionando t√≠tulo e r√≥tulos
plt.title('Vit√≥rias do Santos por Ano')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Vit√≥rias')

# Mostrando o gr√°fico
plt.grid(True)
plt.show()

print(vitorias_por_ano)

```

## Analisando os Gols marcados pelo Santos

```python
# Gols marcados pelo Santos em casa
gols_casa = int(jogos_santos[jogos_santos['Home'] == 'Santos']['HG'].sum())

# Gols marcados pelo Santos fora de casa
gols_fora = int(jogos_santos[jogos_santos['Away'] == 'Santos']['AG'].sum())

print(f'Gols marcados pelo Santos em casa: {gols_casa}')
print(f'Gols marcados pelo Santos fora de casa: {gols_fora}')
print(f'Gols totais: {gols_casa + gols_fora}')
```

Analisando os Gols ao longo dos anos

```python
# Filtrar os jogos do Santos como mandante e visitante
gols_casa = jogos_santos[jogos_santos['Home'] == 'Santos'].groupby(jogos_santos['Date'].dt.year)['HG'].sum()
gols_fora = jogos_santos[jogos_santos['Away'] == 'Santos'].groupby(jogos_santos['Date'].dt.year)['AG'].sum()

# Somar os gols marcados em casa e fora para cada ano
gols_por_ano = gols_casa.add(gols_fora, fill_value=0)

# Exibir a quantidade de gols marcados em cada ano
print(gols_por_ano)

# Plotando o gr√°fico de linha
gols_por_ano.plot(kind='line', marker='o', color='blue', figsize=(10, 6))

# Adicionando t√≠tulo e r√≥tulos
plt.title('Gols Marcados pelo Santos por Ano')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Gols')

# Mostrando o gr√°fico
plt.grid(True)
plt.show()
```
